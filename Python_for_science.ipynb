# Data types and Numpy ndArrays

We know the lists:

```python
l = [ 1,5,3,7,8,9]
print(l)
```

They are usefull, but not very efficient for storing data:

```python
L = range(1000)

%timeit [i**2 for i in L]

import numpy as np
a = np.arange(1000)
%timeit a**2
```

## 1D to nD arrays

```python
print("1D :")
a = np.array([0, 1, 2, 3])
print(a)

s = f"ndim: {a.ndim}, shape: {a.shape}, lenght: {len(a)}"
print(s)

print("2D :")
a = np.array([[0, 1, 2], [3, 4, 5]])
print(a)

s = f"ndim: {a.ndim}, shape: {a.shape}, lenght: {len(a)}"
print(s)

```

# Visualization

To Visualize your arrays, just plot them !
`matplotlib` is the way to go:

```python
import matplotlib.pyplot as plt
Np = 500
a = np.linspace(1,5,Np)  # creat the arrnay [ 1, ... 5] of size Np
b = np.random.randn(Np)  # create Np random data

plt.figure(figsize=(10,3)) # set the size of the plot

plt.subplot(131)
plt.plot(a,b)    # simple line plot
plt.ylabel("the data")   # set the y label


plt.subplot(132)
plt.plot(a,b, 'o', c="green")  # changing the line style and marker
plt.xlabel("the time")   # set the x label

plt.subplot(133)
plt.hist(b)
plt.title("An easy histogram")
```

## Example: special functions

Want a Bessel function ? Search no more: `scipy.special.jn()``

```python
import scipy.special as spec

a = np.linspace(-50,50,500)
b = spec.jn(0,a)

plt.plot(a,b)
plt.title("Bessel function of the first kind of real order and complex argument.")
plt.xlabel("z")
plt.ylabel("$J_0(z)$");
```
